```python
import json
```

```python
with open("C:/Users/rp177/Downloads/IN_category_id.json") as f:
    data = json.load(f)
```

```python
print(type(data))
```

```python
new_data = []
for item in data['items']:
    new_item = {
        'id': item['id'],
        'title': item['snippet']['title'],
        'assignable': item['snippet']['assignable']
    }
    new_data.append(new_item)

# save the cleaned data
with open('C:/Users/rp177/Downloads/IN_category_id_cleaned.json', 'w') as f:
    json.dump(new_data, f, indent=2)

# print cleaned data to check
print(json.dumps(new_data, indent=2))
```
